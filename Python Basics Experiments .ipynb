```python
# define a class
class Dog:
    sound = "bark"  # class attribute
    
#an object is an instance of a class

# Create an object from the class
dog1 = Dog() #dog1 is a object of the class bc it calls the class

# Access the class attribute
print(dog1.sound)

#we now overwrite our class with a prior one
class Dog:
    species = "Canine"  # Class attribute #true for each thing in the class

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age  # Instance attribute
        
        
dog1 = Dog("Buddy", 3) #creates a new object of our new class

print(dog1.name)  # Output: Buddy
print(dog1.species)  # Output: Canine


#now we again redefine our class
class Dog:
    def __init__(self, name, age):  
        self.name = name 
        self.age = age

    def bark(self):  #this is a function that runs when we call the class based on other defs in the class
        print(f"{self.name} is barking!")

# Creating an instance of Dog
dog1 = Dog("Buddy", 3)
dog1.bark() 


#All veriables defined in a class can be defined as either class veriables or instance veriables
#Class veriables are shared accross the class 
#instance veriables are unique to each object 

#example of the diffrence
class Dog:
    # Class variable
    species = "Canine"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

# Create objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

# Access class and instance variables
print(dog1.species)  # (Class variable)
print(dog1.name)     # (Instance variable)
print(dog2.name)     # (Instance variable)

# Modify instance variables
dog1.name = "Max"
print(dog1.name)     # (Updated instance variable)

# Modify class variable
Dog.species = "Feline"
print(dog1.species)  # (Updated class variable)
print(dog2.species)  
```

```python
#Anthor set of examples

class MyClass: #creates a class
    x = 5 #defines a class veriable
    
print (MyClass.x) #calls the class veriable as an object so we can print it

class Person: #defines a new class
    def __init__(self, name, age): 
        self.name = name
        self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)


#we can do a similar thing but when the object is represented as a string we call the final result as ___str___
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}({self.age})"

p1 = Person("John", 36)

print(p1)



#you can also call spacific classes from a class

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def myfunc(self):
        print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.myfunc()



#you can natrally change the name of the self command
class Person:
    def __init__(mysillyobject, name, age): #but it must remain in the same position
        mysillyobject.name = name
        mysillyobject.age = age

    def myfunc(abc):
        print("Hello my name is " + abc.name)

p1 = Person("John", 36)
p1.myfunc()

del p1.age #you can also delete atributes or whole objects

print (p1.age) #now when I call it I get an error
```

```python
x = lambda a : a + 10
print(x(5))  #lambda functions are short and unnamed, this is useful for making nested functions shorter

x = lambda a, b : a * b
print(x(5, 6)) #we can have lambda functions with multiple veriables

def myfunc(n): #for instance we have this code which is made more efficant by the use of a lambda function
    return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```
