# 3-D Random Walk

```python
# Your program here
import numpy as np
import matplotlib.pyplot as plt #imports needed funcition

L = 101 #defines the width and hight of our area   #L MUST BE ODD

Step = 10000 #the number of steps we will be taking

Step_Diretion = np.random.randint (low = 0, high = 6, size = Step) #gives us a random list four possible output numbers
 #we let 0 to down, one be up, two be left and three be right, 4 to be out and 5 to be in, giving us all possible directions

x_pos = 50
y_pos = 50  #we start our particle at the centre of the grid
z_pos = 50

x_pos_list = np.zeros (Step) #creates a saveplace for each step so we can recall the results
y_pos_list = np.zeros (Step)
z_pos_list = np.zeros (Step)

for k in range(len(Step_Diretion)): #for each random number we generated, for each step
    if Step_Diretion [k] == 0 and y_pos != 0: #if down we subtract one from y
        y_pos -= 1
    if Step_Diretion [k] == 1 and y_pos != L - 1: #if up we add one to y
        y_pos += 1
    if Step_Diretion [k] == 2 and x_pos != 0: #if left we subract one from x
        x_pos -= 1
    if Step_Diretion [k] == 3 and x_pos != L - 1: #if right we add one to x
        x_pos += 1 #Note we also have a rule that says we do nothing if we are at a boundary and told to go further into it
    if Step_Diretion [k] == 4 and z_pos != 0: #if down we subtract one from y
        z_pos -= 1
    if Step_Diretion [k] == 5 and z_pos != L - 1: #if down we subtract one from y
        z_pos += 1
    x_pos_list[k] += x_pos
    y_pos_list[k] += y_pos #we save our results into a our zeros list
    z_pos_list[k] += z_pos

fig, ax = plt.subplots (subplot_kw = {"projection" : "3d"})
P = ax.scatter (x_pos_list,y_pos_list,z_pos_list, c = np.arange (Step),cmap = "cool") #plot our results
ax.set_xlabel ("x position") #lables our axis
ax.set_ylabel ("y position")
ax.set_zlabel ("z position")
ax.set_title (f"100 x 100 x 100 grid 2_D Random Walk Simultion for {Step} steps") #titles
fig.colorbar (P, label = "number of steps", shrink  = 0.6, location = "bottom")
plt.show () #tells Python to show our graph
```
